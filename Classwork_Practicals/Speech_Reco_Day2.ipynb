```python
!pip install librosa
import librosa
import librosa.display
from sklearn.preprocessing import minmax_scale
import IPython.display as ipd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
ipd.Audio('/content/angry_woman.wav')
```

```python
ipd.Audio('/content/angry_man.wav')
```

```python
x, sr =librosa.load('/content/angry_woman.wav')
plt.figure(figsize=(20,5))
plt.title('Waveplot of Angry Woman')
librosa.display.waveshow(x,sr=sr)
plt.show()
```

```python
x1, sr =librosa.load('/content/angry_man.wav')
plt.figure(figsize=(20,5))
plt.title('Waveplot of Angry Man')
librosa.display.waveshow(x1,sr=sr)
plt.show()
```

```python
#Spectogram
plt.figure(figsize=(20,5))
plt.title('Speechshow of Angry Woman')
X=librosa.stft(x)

Xdb=librosa.amplitude_to_db(abs(X))
librosa.display.specshow(Xdb,sr=sr,x_axis='time',y_axis='log')
plt.show()

#Spectogram
plt.figure(figsize=(20,5))
plt.title('Speechshow of Angry Man')
X1=librosa.stft(x1)

Xdb1=librosa.amplitude_to_db(abs(X1))
librosa.display.specshow(Xdb1,sr=sr,x_axis='time',y_axis='log')
plt.show()
```

```python
# Xero cross rating - used for noise identifcation

zero_crossings=librosa.zero_crossings(x)
print("Sum of Zero Crossing", zero_crossings.sum())
plt.figure(figsize=(20,5))
plt.title('Zero Crossing Rate of Angry Woman')
zcrs=librosa.feature.zero_crossing_rate(x)
plt.plot(zcrs[0])
plt.show()

zero_crossings1=librosa.zero_crossings(x1)
print("Sum of Zero Crossing", zero_crossings1.sum())
plt.figure(figsize=(20,5))
plt.title('Zero Crossing Rate of Angry Man')
zcrs1=librosa.feature.zero_crossing_rate(x1)
plt.plot(zcrs1[0])
plt.show()
```

```python
#Spectral Centroids

from sklearn.preprocessing import minmax_scale
def normalize(x,axis = 0):
  return minmax_scale(x,axis=axis)
```

```python
plt.figure(figsize=(20,5))
plt.title('Spectral Cantroids of angry woman')
spectral_centroids=librosa.feature.spectral_centroid(y=x,sr=sr)[0]
frames=range(len(spectral_centroids))
t=librosa.frames_to_time(frames)
librosa.display.waveshow(x,sr=sr,alpha=0.75)
plt.plot(t,normalize(spectral_centroids),color='r')
plt.show()
```

```python
#spectral Roll off
plt.figure(figsize=(20,5))
plt.title('Spectral Roll off of angry man')
spectral_rolloff=librosa.feature.spectral_rolloff(y=x1,sr=sr)[0]
frames=range(len(spectral_rolloff))
t=librosa.frames_to_time(frames)
librosa.display.waveshow(x1,sr=sr,alpha=0.4)
plt.plot(t,normalize(spectral_rolloff),color='r')
plt.show()
```
