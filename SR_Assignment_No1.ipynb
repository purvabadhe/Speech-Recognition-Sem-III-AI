# Q. 2 Analyze that audio using librosa waveplot and spectrogram.


```python
x,sr=librosa.load(audio_path)
```

```python
import matplotlib.pyplot as plt
import librosa.display
```

```python
#Visualize the audio

plt.figure(figsize=(10,5))
librosa.display.waveshow(x,sr=sr)
```

```python
#Spectogram

X= librosa.stft(x) #short term fourier transform
Xdb=librosa.amplitude_to_db(abs(X)) #Amplitude to frequency at given time
```

```python
plt.figure(figsize=(14,5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time',y_axis='hz')
plt.colorbar()
```

# Q. 3 Comment on the output

Waveplot:
The waveplot displays the amplitude of the audio signal over time for the provided text. Since the text consists of a list of words, the waveplot will show the overall shape of the audio signal as it changes during the pronunciation of these words. The x-axis represents time, and the y-axis represents the amplitude (loudness) of the audio at different time points.

The waveplot will likely show periodic patterns or fluctuations in amplitude as the words are spoken one after the other. The pauses between words will also be visible as flat regions in the waveform.

The amplitude variations may differ depending on the length and complexity of the words. Longer words may have more extended segments with varying amplitudes.

Spectrogram:
The spectrogram displays the frequency content of the audio signal over time for the given text. Each word in the text will contribute to the spectral characteristics seen in the spectrogram. Different sounds from the words will be represented by distinct frequency patterns, which can be observed as vertical bands in the spectrogram.

The spectrogram will help identify the frequency patterns associated with each word, and the transitions between words may be observable as shifts in the frequency content.
